```python
import pandas as pd
import numpy as np



movies = pd.read_csv("../data/raw/movies.csv")
rating = pd.read_csv("../data/raw/ratings.csv")
tags = pd.read_csv("../data/raw/tags.csv")


print(movies.head())
print(rating.head())
print(tags.head())
```

```python
print("movies data shape", movies.shape)
print("rating data shape", rating.shape)
print("tags data shape", tags.shape)

print("\n")

print('missing values movie', movies.isnull().sum())\
    
print("\n")

print('missing values rating', rating.isnull().sum())

print("\n")

print('missing values tags', tags.isnull().sum())

```

```python
tags.dropna(inplace=True)
```

```python
import matplotlib.pyplot as plt 
import os

movies_rating =  pd.merge(movies,rating,on='movieId')
top_rated_movies= movies_rating['title'].value_counts().head(10)

plt.figure(figsize = (12,12))
top_rated_movies.plot(kind='bar', color='black')
plt.title('top 10 most rated movies')
plt.xlabel("Movie title")
plt.ylabel("rating")
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('../reports/top_10_rated_movies.png', format='png')
plt.show()
```
